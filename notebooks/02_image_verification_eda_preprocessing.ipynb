```python
def image_exists(tweet_id):
    tweet_id = str(tweet_id)

    for base_dir in [DEV_IMG_2015, TEST_IMG_2015]:
        if not os.path.exists(base_dir):
            continue

        # iterate over event folders
        for event in os.listdir(base_dir):
            event_dir = os.path.join(base_dir, event)
            if not os.path.isdir(event_dir):
                continue

            for fname in os.listdir(event_dir):
                if fname.startswith(tweet_id + "_"):
                    return os.path.join(event_dir, fname)

    return None
```

```python
img_df = pd.read_csv(
    os.path.join(BASE_PATH, "tweets_images.txt"),
    sep="\t",
    header=None,
    names=["tweet_id", "image_id", "label"],
    engine="python",
    on_bad_lines="skip"
)

img_df.dropna(inplace=True)
img_df.drop_duplicates(inplace=True)

print("Dataset size after cleaning:", len(img_df))
img_df.head()
```
